```python
import os

from dotenv import load_dotenv
load_dotenv()

os.environ['OPENAI_API_KEY']=os.getenv('OPENAI_API_KEY')

# Langsmith tracking

os.environ['LANGCHAIN_API_KEY']=os.getenv('LANGCHAIN_API_KEY')
os.environ['LANGCHAIN_PROJECT']=os.getenv('LANGCHAIN_PROJECT')

os.environ['LANGCHAIN_TRACING_V2']='True'
 
 
```

```python
from langchain_openai import ChatOpenAI

llm=ChatOpenAI(model='gpt-4o')
print(llm)
```

```python
# Input and get response from LLM

llm.invoke("What is Generative AI")
```

```python
# ChatPrompt Template

from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate.from_messages(
    [
        ("system", "You are expert AI engineer. Provide me answers based on the questions"),
        ("user", "{input}")
    ]
)

prompt
```

```python
# Chain -> Combines multi things

# It will show it in content variable

chain=prompt|llm
response=chain.invoke({"input": "Can you tell me about langsmith?"})

print(response)
```

```python
# StrOutputParser

# It will not show in content variable

from langchain_core.output_parsers import StrOutputParser

output_parser=StrOutputParser()
chain=prompt|llm|output_parser

response=chain.invoke({"Can you tell me about LangSmith?"})

print(response)
```
